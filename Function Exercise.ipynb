```python
#5. Write a function which will take list/multiple list as an input and return the
#concatenation of list as an output
```

```python
def concat_list(*x):
    y= []
    for i in x:
        for j in i:
            y.append(str(j))
    s="".join(y)
    return s
            
```

```python
import platform,socket,re,uuid,json,psutil,logging

def getSystemInfo():
    try:
        info={}
        info['platform']=platform.system()
        info['platform-release']=platform.release()
        info['platform-version']=platform.version()
        info['architecture']=platform.machine()
        info['hostname']=socket.gethostname()
        info['ip-address']=socket.gethostbyname(socket.gethostname())
        info['mac-address']=':'.join(re.findall('..', '%012x' % uuid.getnode()))
        info['processor']=platform.processor()
        info['ram']=str(round(psutil.virtual_memory().total / (1024.0 **3)))+" GB"
        return json.dumps(info)
    except Exception as e:
        logging.exception(e)

json.loads(getSystemInfo())
```

https://towardsdatascience.com/5-python-open-source-tools-to-extract-text-and-tabular-data-from-pdf-files-9359cdde6b56

```python
def read_pdf(x):
    from PyPDF4 import PdfFileReader

    # creating a pdf file object
    pdfObject = open(x, 'rb')

    # creating a pdf reader object
    pdfReader = PdfFileReader(pdfObject)

    # Extract and concatenate each page's content
    text=''
    for i in range(0,pdfReader.numPages):
        # creating a page object
        pageObject = pdfReader.getPage(i)
        # extracting text from page
        text += pageObject.extractText()
    print(text)
```

```python
read_pdf('pdf_test2.pdf')
```

```python
def get_ip():
    import socket

    s = socket.socket(socket.AF_INET, socket.SOCK_DGRAM)
    s.connect(('8.8.8.8', 1))  # connect() for UDP doesn't send packets
    local_ip_address = s.getsockname()[0]
    return local_ip_address
```

```python
#20.Merging  pdf available in folder
```

```python
def merge_pdf(x):
    import os
    from PyPDF4 import PdfFileMerger

    source_dir = x
    
    merger = PdfFileMerger()

    for item in os.listdir(source_dir):
        if item.endswith('pdf') and item.startswith('pdf_test') :
            #print(item)
            merger.append(source_dir + item)

    merger.write(source_dir + 'Complete.pdf')       
    merger.close()
```

```python
merge_pdf('/Users/saumyapandey/Documents/ineuron/')
```

```python
#just updating two or more files is also possible
from PyPDF4 import PdfFileMerger

pdfs = ['pdf_test2.pdf','pdf_test3.pdf']

merger = PdfFileMerger()

for pdf in pdfs:
    merger.append(pdf)

merger.write("result.pdf")  # writing line
merger.close()
```
